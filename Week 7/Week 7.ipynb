csv 파일에 있는 정보를 접속하려면, iloc 기능을 이용하여 일반적으로 익숙한 리스트 자료 접속과 비슷하게 이용할 수 있습니다!


```python
apple.iloc[4, 0] # 5번째 행 (row), 첫번째 열 (column)
```

```python
apple.iloc[1:4, 3] #2nd to 4th row, 4th column
```

```python
apple.iloc[1:4, 0:4] #2nd to 4th row, 1st to 4th column
```

```python
apple.iloc[-4:, -1]
```
